```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
data=pd.read_csv('/content/drive/MyDrive/Heart_disease_cleveland_new.csv')
```

```python
data.tail()
```

```python
data.drop(columns=['ca','thal'], inplace=True)
```

```python
data[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','target']].hist(figsize=(13, 8))
plt.tight_layout()
plt.show()
```

```python
import seaborn as sb
cormat=data.corr()
top_corr_feature=cormat.index
plt.figure(figsize=(10,8))
g=sb.heatmap(data[top_corr_feature].corr(),annot=True,cmap="RdYlGn")
```

```python
data=pd.get_dummies(data,columns=['sex','cp','fbs','restecg','exang','slope'])
```

```python
from sklearn.preprocessing import MinMaxScaler
MinMaxScaler = MinMaxScaler()
column_scale=['age','trestbps','chol','thalach','oldpeak']
data[column_scale]=MinMaxScaler.fit_transform(data[column_scale])
```

```python
data.tail()
```

```python
heart_disease_count=data['target'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(heart_disease_count,labels=heart_disease_count.index,autopct='%1.1f%%',startangle=100)
plt.title('Distribution of Heart Disease')
plt.axis('equal')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X=data.drop('target',axis=1)
y=data['target']
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import pandas as pd

X_train1, X_test, y_train1, y_test = train_test_split(X, y, test_size=0.15, random_state=28)

# Print class distribution before SMOTE
print("Before SMOTE:")
print(y_train1.value_counts())

# Apply SMOTE to the training data
smote = SMOTE(random_state=28)
X_train, y_train = smote.fit_resample(X_train1, y_train1)

# Print class distribution after SMOTE
print("\nAfter SMOTE:")
print(pd.Series(y_train).value_counts())
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, confusion_matrix,roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.naive_bayes import GaussianNB
# Define individual classifiers

logistic_regression = LogisticRegression()
random_forest = RandomForestClassifier(n_estimators=95,random_state=28)
xgboost = xgb.XGBClassifier(n_estimators=95,random_state=28)
knn = KNeighborsClassifier(n_neighbors=7)
naive_bayes = GaussianNB()
svm = SVC(probability=True,random_state=28)
adaboost= AdaBoostClassifier(random_state=28)

# Train individual classifiers
logistic_regression.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
xgboost.fit(X_train, y_train)
knn.fit(X_train, y_train)
naive_bayes.fit(X_train, y_train)
svm.fit(X_train, y_train)
adaboost.fit(X_train, y_train)

# Make predictions on the test set
y_pred_logistic = logistic_regression.predict(X_test)
y_pred_random_forest = random_forest.predict(X_test)
y_pred_xgboost = xgboost.predict(X_test)
y_pred_knn = knn.predict(X_test)
y_pred_naive_bayes = naive_bayes.predict(X_test)
y_pred_svm = svm.predict(X_test)
y_pred_adaboost = adaboost.predict(X_test)

# Calculate individual classifier performance
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
accuracy_xgboost = accuracy_score(y_test, y_pred_xgboost)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_naive_bayes = accuracy_score(y_test, y_pred_naive_bayes)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
accuracy_adaboost=accuracy_score(y_test,y_pred_adaboost)

# Print individual classifier performance
print("Accuracy Logistic Regression:",  "{:.2%}".format(accuracy_logistic))
print("Accuracy Random Forest:", "{:.2%}".format(accuracy_random_forest))
print("Accuracy XGBoost:", "{:.2%}".format(accuracy_xgboost))
print("Accuracy KNN:", "{:.2%}".format(accuracy_knn))
print("Accuracy Naive Bayes:", "{:.2%}".format(accuracy_naive_bayes))
print("Accuracy SVM:", "{:.2%}".format(accuracy_svm))
print("Accuracy adaboost:", "{:.2%}".format(accuracy_adaboost))
```

```python
precision_logistic = precision_score(y_test, y_pred_logistic,average='macro')
precision_random_forest = precision_score(y_test, y_pred_random_forest,average='macro')
precision_XGBoost = precision_score(y_test, y_pred_xgboost, average='macro')
precision_KNN = precision_score(y_test, y_pred_knn, average='macro')
precision_naive_bayes = precision_score(y_test, y_pred_naive_bayes, average='macro')
precision_svm = precision_score(y_test, y_pred_svm, average='macro')
precision_adaboost = precision_score(y_test, y_pred_adaboost, average='macro')

print("Precision Logistic Regression:",  "{:.2%}".format(precision_logistic))
print("Precision random forest:",  "{:.2%}".format(precision_random_forest))
print("Precision xgboost:",  "{:.2%}".format(precision_XGBoost))
print("Precision knn:",  "{:.2%}".format(precision_KNN))
print("Precision naive bayes:",  "{:.2%}".format(precision_naive_bayes))
print("Precision svm:",  "{:.2%}".format(precision_svm))
print("Precision adaboost:",  "{:.2%}".format(precision_adaboost))
```

```python
recall_logistic = recall_score(y_test, y_pred_logistic,average='macro')
recall_random_forest = recall_score(y_test, y_pred_random_forest,average='macro')
recall_XGBoost = recall_score(y_test, y_pred_xgboost,average='macro')
recall_KNN = recall_score(y_test, y_pred_knn,average='macro')
recall_naive_bayes = recall_score(y_test, y_pred_naive_bayes,average='macro')
recall_svm = recall_score(y_test, y_pred_svm,average='macro')
recall_adaboost = recall_score(y_test, y_pred_adaboost,average='macro')

print("recall logistic:", "{:.2%}".format(recall_logistic))
print("recall random:", "{:.2%}".format(recall_random_forest))
print("recall xgboost:", "{:.2%}".format(recall_XGBoost))
print("recall knn:", "{:.2%}".format(recall_KNN))
print("recall naive bayes:", "{:.2%}".format(recall_naive_bayes))
print("recall svm:", "{:.2%}".format(recall_svm))
print("recall adaboost:", "{:.2%}".format(recall_adaboost))
```

```python
f1_logistic = f1_score(y_test, y_pred_logistic, average='macro')
f1_random_forest = f1_score(y_test, y_pred_random_forest, average='macro')
f1_xgboost = f1_score(y_test, y_pred_xgboost, average='macro')
f1_knn = f1_score(y_test, y_pred_knn, average='macro')
f1_naive_bayes = f1_score(y_test, y_pred_naive_bayes, average='macro')
f1_svm = f1_score(y_test, y_pred_svm, average='macro')
f1_adaboost = f1_score(y_test, y_pred_adaboost, average='macro')

# Print F1 scores for each model
print("F1 score for logistic regression:", "{:.2%}".format(f1_logistic))
print("F1 score for random forest:", "{:.2%}".format(f1_random_forest))
print("F1 score for XGBoost:", "{:.2%}".format(f1_xgboost))
print("F1 score for KNN:", "{:.2%}".format(f1_knn))
print("F1 score for naive Bayes:", "{:.2%}".format(f1_naive_bayes))
print("F1 score for SVM:", "{:.2%}".format(f1_svm))
print("F1 score for AdaBoost:", "{:.2%}".format(f1_adaboost))
```

```python
from sklearn.ensemble import VotingClassifier

voting_classifier = VotingClassifier(
    estimators=[
    ('Logistic Regression',logistic_regression),
    ('xgboost', xgboost),
    ('svm', svm),
    ('random_forest',random_forest),
    ('adaboost',adaboost),
    ('KNN', knn),
    ('Naive Bayes', naive_bayes),

], voting='hard')
start_time = time.time()
voting_classifier.fit(X,y)
y_pred_voting = voting_classifier.predict(X_test)
end_time = time.time()
execution_time = end_time - start_time
print("Execution time:", "{:.2f}".format(execution_time), "seconds")

accuracy = accuracy_score(y_test, y_pred_voting)
print("Voting Classifier Accuracy:", "{:.2%}".format(accuracy))
precision_voting = precision_score(y_test, y_pred_voting,average='macro')
print("Voting Precision:", "{:.2%}".format(precision_voting))


recall_voting = recall_score(y_test, y_pred_voting,average='macro')
print("Voting Recall:", "{:.2%}".format(recall_voting))

f1_voting = f1_score(y_test, y_pred_voting,average='macro')
print("Voting F1-score:", "{:.2%}".format(f1_voting))
```

```python
np.random.seed(28)
from sklearn.ensemble import StackingClassifier
base_models = [
    ('random_forest',random_forest),
    ('svm',svm),
    ('Naive Bayes', naive_bayes),
    ('xgboost',xgboost),
    ('adaboost',adaboost),
    ('KNN', knn),
    #('lr',logistic_regression)
    ]

meta_learner = logistic_regression

stacking_classifier = StackingClassifier(
    estimators=base_models,final_estimator=meta_learner)
start_time = time.time()
stacking_classifier.fit(X, y)

y_pred_stacking = stacking_classifier.predict(X_test)
end_time = time.time()
execution_time = end_time - start_time
print("Execution time:", "{:.2f}".format(execution_time), "seconds")

accuracy = accuracy_score(y_test, y_pred_stacking)
print("Stacking Accuracy:","{:.2%}".format(accuracy))
precision_stacking = precision_score(y_test, y_pred_stacking,average='macro')
print("stacking Precision:", "{:.2%}".format(precision_stacking))


recall_stacking = recall_score(y_test, y_pred_stacking,average='macro')
print("stacking Recall:", "{:.2%}".format(recall_stacking))

f1_stacking = f1_score(y_test, y_pred_stacking,average="macro")
print("stacking F1-score:", "{:.2%}".format(f1_stacking))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Define function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# Plot confusion matrix for each model
plot_confusion_matrix(y_test, y_pred_logistic, 'Confusion Matrix - Logistic Regression')
```
